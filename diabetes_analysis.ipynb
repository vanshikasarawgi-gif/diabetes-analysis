<h1> Diabetes Data Analysis <h1>

<h3>This project analyzes the diabetes dataset using Pandas, NumPy, Seaborn and Matplotlib.<h3>


<h3> importing libraries and loading the file <h3>


```python
import pandas as pd
df = pd.read_csv('data/diabetes.csv')
```

<h3> Checking for null values <h3>

```python
df.isnull().sum()
```

<h3> Average glucose level for each age <h3>

```python
df.groupby('Age')['Glucose'].mean().sort_values(ascending=True)
```

<h3> Adding column of average glucose by age<h3>

```python
df['Avg_Glucose_By_Age'] = df.groupby('Age')['Glucose'].transform('mean')
```

```python
df
```

<h3> Plotting the average glucose value by age using Matplotlib <h3>

```python
import matplotlib.pyplot as plt
age_glucose = df.groupby('Age')['Glucose'].mean()
plt.figure(figsize=(10, 5))
plt.plot(age_glucose.index, age_glucose.values, marker='o', color='blue')
plt.title('Average Glucose Level by Age')
plt.xlabel('Age')
plt.ylabel('Average Glucose Level')
plt.grid(True)
plt.show()
```

<h3> Grouping BMI into ranges <h3>

```python
bins = [0,18.5,24.9,29.5,50]
labels = ['Underweight','Normal','Overweight','Obese']
df['BMI_category'] = pd.cut(df['BMI'], bins=bins,labels=labels)
```

```python
df
```

```python
df.groupby('BMI_category',observed=True)['Glucose'].mean().sort_values(ascending=True)
```

<h3> Plotting average glucose levels by BMI Category <h3>

```python
import matplotlib.pyplot as plt
bmi_glucose = df.groupby('BMI_category',observed=True)['Glucose'].mean()
plt.figure(figsize=(10, 5))
plt.bar(bmi_glucose.index, bmi_glucose.values, color='red')
plt.title('Average Glucose Level by BMI category')
plt.xlabel('BMI category')
plt.ylabel('Average Glucose Level')
plt.grid(True)
plt.show()
```

<h3> Correlation between columns <h3>

```python
df.corr(numeric_only=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = df.corr(numeric_only=True)

# plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Diabetes Dataset')
plt.show()
```

<h3> Relationship between BMI and glucose <h3>

```python
plt.scatter(df['Glucose'], df['BMI'], c=df['Outcome'], cmap='coolwarm')
plt.title('Glucose vs BMI (Colored by Diabetes Outcome)')
plt.xlabel('Glucose Level')
plt.ylabel('BMI')
plt.colorbar(label='Outcome (0 = No Diabetes, 1 = Diabetes)')
plt.show()
```

<h3> Comparing blood pressure for people with diabetes and without diabetes <h3>

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.scatter(df['BloodPressure'], df['Outcome'], alpha=0.5, color='blue')
plt.xlabel("Blood Pressure")
plt.ylabel("Diabetes Outcome")
plt.title("Scatter: Blood Pressure vs Diabetes Outcome")
plt.yticks([0,1], ["No Diabetes", "Diabetes"])
```

<h3> Diabetes pedigree function vs Age<h3>

```python
import matplotlib.pyplot as plt

plt.scatter(df['Age'], df['DiabetesPedigreeFunction'], alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Diabetes Pedigree Function")
plt.title("Scatter: Age vs Diabetes Pedigree Function")
plt.show()

```

<h3> Insulin vs Glucose <h3>

```python
insulin_level = df.groupby('Glucose')['Insulin'].mean()

plt.figure(figsize=(10, 5))
plt.plot(insulin_level.index, insulin_level.values, marker='o', color='blue')
plt.title('Average Insulin Level by Glucose')
plt.xlabel('Glucose')
plt.ylabel('Average Insulin')
plt.grid(True)
plt.show()
```
